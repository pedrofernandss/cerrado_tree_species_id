```python
import shutil
import re
from pathlib import Path
```

```python
def extrair_id_unico(nome_arquivo):
    match = re.search(r"(DJI_\d{14}_\d{4})", nome_arquivo)
    if match:
        return match.group(1) 
    return None
```

```python
path_labels_source = Path("/mnt/sdb-seagate/graduacao/datasets/projeto_cerrado/fotos-rotuladas/rgb/labels")
path_images_source = Path("/mnt/sdb-seagate/graduacao/datasets/projeto_cerrado/2024-08-28/rgb-ndre-imgs")
path_output = Path("/mnt/sdb-seagate/graduacao/datasets/projeto_cerrado/fotos-rotuladas/rgb-ndre")
```

```python
(path_output / "images").mkdir(parents=True, exist_ok=True)
(path_output / "labels").mkdir(parents=True, exist_ok=True)
```

```python
dict_imagens = {}
lista_imgs = list(path_images_source.glob("*.[tT][iI][fF]")) 

print(f"üìÇ Indexando {len(lista_imgs)} imagens...")
for img in lista_imgs:
    id_img = extrair_id_unico(img.name)
    if id_img:
        dict_imagens[id_img] = img
    else:
        print(f"‚ö†Ô∏è Aviso: Imagem sem ID reconhec√≠vel: {img.name}")

lista_labels = list(path_labels_source.glob("*.txt"))
print(f"üìÇ Processando {len(lista_labels)} labels...")

matches = 0
erros = 0

for lbl in lista_labels:
    id_lbl = extrair_id_unico(lbl.name)
    
    if id_lbl and id_lbl in dict_imagens:
        caminho_img = dict_imagens[id_lbl]
        
        try:
            shutil.copy(caminho_img, path_output / "images" / caminho_img.name)

            novo_nome_label = caminho_img.stem + ".txt"
            shutil.copy(lbl, path_output / "labels" / novo_nome_label)
            
            matches += 1
            
        except Exception as e:
            print(f"Erro ao copiar {id_lbl}: {e}")
            erros += 1

print("\n" + "="*40)
print(f"Total Imagens na Fonte: {len(lista_imgs)}")
print(f"Total Labels na Fonte:  {len(lista_labels)}")
print(f"Pares Formados:         {matches}")
print(f"Pasta Final:            {path_output}")
print("-" * 40)
```
